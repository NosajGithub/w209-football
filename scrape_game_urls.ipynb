```python
import urllib
from bs4 import BeautifulSoup
import pandas as pd
import csv
from random import randint
from time import sleep
import lxml

def get_game_urls(team, year):
    """
    Scrapes one page of game urls
    """

    # Scrape game urls
    content = urllib.urlopen("http://www.pro-football-reference.com/teams/" + team + "/" + year + ".htm")
    s = content.read()
    # s = open("vikings.html")

    soup = BeautifulSoup(s, "lxml")

    table = soup.find("table", { "id" : "team_gamelogs" }).find("tbody")
    # print table.prettify()

    game_urls = []
    for row in table.findAll("tr"):
        for num, data in enumerate(row.findAll("td")):
            if num == 3:
                if data.a is not None:
                    game_urls.append(data.a.get("href"))
#                     print data.a.get("href")

    return game_urls

teams = ["crd", "atl", "rav", "buf","car","chi","cin","cle","dal","den","det","gnb","htx","clt","jax","kan","mia","min",
           "nwe","nor","nyg","nyj","rai","phi","pit","sdg","sfo","sea","ram","tam","oti","was"]
# teams = ["crd", "atl"]
years = ['2009','2010','2011','2012','2013','2014']
# years1 = ['2009','2010','2011']
# years2 = ['2012','2013','2014']
team_years = [(x,y) for x in teams for y in years]

## Scrape each of the pages
game_urls = []

for a,b in team_years:
    print a, b
    sleep(randint(1,3))
    sleep(randint(1,3))    
    game_urls += get_game_urls(a,b)

game_urls
```
