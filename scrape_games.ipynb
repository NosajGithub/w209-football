```python
import urllib
from bs4 import BeautifulSoup
import pandas as pd
import csv
from random import randint
from time import sleep
import lxml

def getGame(url):
    """
    Scrapes one page of injury information
    """
    surface = ""
    roof = ""
    start_time = ""
    weather = ""
    
    # Scrape injury information
    content = urllib.urlopen("http://www.pro-football-reference.com/" + url)
    s = content.read()
#     s = open("game.htm")

    soup = BeautifulSoup(s, "lxml")

    game = soup.find("div", { "id" : "div_game_info" })

    for num, heading in enumerate(game.findAll("td")):
        if heading.text == "Surface":
            surface = game.findAll("td")[num+1].text
        if heading.text == "Roof":
            roof = game.findAll("td")[num+1].text
        if heading.text == "Start Time (ET)":
            start_time = game.findAll("td")[num+1].text
        if heading.text == "Weather":
            weather = game.findAll("td")[num+1].text

    return {'surface':surface,
           'roof':roof,
           'weather':weather,
           'start_time':start_time}
```

```python
df = pd.read_csv("game_urls.csv")
```

```python
game_url = []
roof = []
start_time = []
surface = []
weather = []

# urls = list(df.boxscores[0:len(df.boxscores)/2])
urls = list(df.boxscores[len(df.boxscores)/2:])

for num, url in enumerate(urls):
    print num
    sleep(randint(2,4))
    result = getGame(url)
    game_url.append(url)
    roof.append(result['roof'])
    start_time.append(result['start_time'])
    surface.append(result['surface'])
    weather.append(result['weather'])
    
dataset = pd.DataFrame({'game_url': game_url, 
                       'roof': roof, 
                       'start_time': start_time, 
                       'surface': surface, 
                       'weather': weather})

# dataset.to_csv("game_info_1.csv")
dataset.to_csv("game_info_2.csv")
dataset
```
